```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
data = pd.read_csv('salary_dataset.csv')
data.head()
```

```python
data.drop('Company Name',axis = 1, inplace= True)
data.head()
```

```python
data['Job Title'].value_counts()
```

```python
data.rename(columns = {'Job Title':'Job_Title', 'Salaries Reported':'Salaries_Reported'}, inplace = True)
data.head()
```

```python
data.Job_Title = LabelEncoder().fit_transform(data.Job_Title)
data.head()
```

```python
data.duplicated().sum()
```

```python
data.isna().sum()
```

```python
data.dtypes
```

```python
data['Salaries_Reported'].dtype
```

```python
data['Location'].unique()
```

```python
data.Location = LabelEncoder().fit_transform(data.Location)
data.head()
```

```python
data['Salary'] = data['Salary'].apply(lambda x : x[1:])
data['Salary'] = data['Salary'].str.replace(',','')
data.tail()
```
