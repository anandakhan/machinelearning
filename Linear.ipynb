```python
fig, axs = plt.subplots( 1, 3, sharey = True)
df.plot(kind='scatter',x='TV',y='sales',ax = axs[0],color='red', figsize=(16,5))
df.plot(kind='scatter',x='radio',y='sales',color= 'blue',ax=axs[1])
df.plot(kind='scatter',x='newspaper',y='sales', color = 'green',ax=axs[2])
```

```python
feature_cols = ['TV']
X = df[feature_cols]
y = df['sales']
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)

print(lr.intercept_)
print(lr.coef_)
```

```python
import numpy as np
np.random.seed(145)
nums = np.random.rand(len(df))
mask_large = nums >0.5
df['Scale'] = 'small'
df.loc[mask_large,'Scale'] = 'large'
df.head()
```

```python
df['Islarge'] = df.Scale.map({'small': 0,'large': 1})
df.head()
```

```python
feature_cols = ['TV','radio','newspaper','Islarge']
X = df[feature_cols]
y = df['sales']
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)

i = 0
for col in feature_cols:
  print('the coefficient of ',col, 'is:',lr.coef_[i])
  i = i+1
```

```python
np.random.seed(1456)
nums = np.random.rand(len(df))
mask_sundarban = (nums > 0.33) & (nums < 0.66)
mask_urban = nums > 0.66
df['Targeted_Geography'] = 'rural'
df.loc[mask_sundarban,'Targeted_Geography'] = 'sundarban'
df.loc[mask_urban,'Targeted_Geography'] = 'urban'
df.head()

```

```python
area_dummies = pd.get_dummies(df['Targeted_Geography'],prefix = 'Targeted_Geography').iloc[:,1:]
df = pd.concat([df,area_dummies],axis=1)
df.head()
```

```python
feature_cols = ['TV','radio','newspaper','Targeted_Geography_sundarban','Targeted_Geography_urban']
X = df[feature_cols]
y = df['sales']
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)

print(feature_cols)
print(lr.coef_)
```
