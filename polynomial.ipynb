```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Position_Salaries.csv')
data.head() 
```

```python
x = data.iloc[:,1:2].values
x
```

```python
y = data.iloc[:,2].values
y
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

```python
plt.scatter(x,y,color='green')
plt.plot(x,lin_reg.predict(x),color='red')
plt.title('Linear Regression')
plt.xlabel('x->level')
plt.ylabel('y->salary')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_fe = PolynomialFeatures(degree=2)
poly1 = poly_fe.fit_transform(x)
print(poly1)
poly_fe.fit(poly1,y)

```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(poly1,y)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg2.predict(poly1),color='green')
plt.title('Linear Regression of 2nd degree')
plt.xlabel('x->level')
plt.ylabel('y-> salary')
plt.show()
```

```python
poly_fe1 = PolynomialFeatures(degree=4)
poly2 = poly_fe1.fit_transform(x)
poly_fe1.fit(poly2,y)
lin_reg3 = LinearRegression()
lin_reg3.fit(poly2,y)




```

```python
plt.scatter(x,y,color='green')
plt.plot(x,lin_reg3.predict(poly2),color='red')
plt.title('Linear Regression of 4th degree')
plt.xlabel('x->level')
plt.ylabel('y-> salary')

plt.show()
```
