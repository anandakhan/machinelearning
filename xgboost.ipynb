```python
import pandas as pd
import numpy as np
import pickle
import xgboost as xgb
from sklearn import datasets
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
```

```python
data = pd.read_csv('pima-indians-diabetes.csv')
data.head(6)
```

```python
data.isna().sum()
```

```python
X = data.drop(columns = 'Is Diabetic')
y = data['Is Diabetic']
```

```python
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X_scaled, y, test_size = 0.3 , random_state = 244)
```

```python
model = XGBClassifier(objective ='binary:logistic')
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_train)
accuracy = accuracy_score(y_train, y_pred)
accuracy
```

```python
y_pred = model.predict(x_test)
print(y_pred)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
param_grid = {'learning_rate':[1,0.5,0.1,0.01,0.001],'max_depth':[3,5,10,20],'n_estimator':[10,50,100,200]}
```

```python
grid = GridSearchCV(XGBClassifier(objective='binary:logistic'),param_grid, verbose = 3)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
new_model = XGBClassifier(learning_rate= 0.1, max_depth= 3, n_estimator= 10)
new_model.fit(x_train,y_train)
```

```python
y_pred = new_model.predict([[5,116,74,0,0,25.6,0.201,30]])
print(y_pred)
#accuracy = accuracy_score(y_test,y_pred)
#accuracy
```

```python
filename = 'xgboost_model.pickle'
pickle.dump(new_model,open(filename,'wb'))
loaded_model = pickle.load(open(filename,'rb'))
```

```python
filename_scaler ='scaler_model.pickle'
pickle.dump(sc,open(filename_scaler,'wb'))
scaler_model = pickle.load(open(filename_scaler,'rb'))
```
